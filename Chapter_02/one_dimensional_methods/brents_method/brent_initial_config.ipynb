```julia
using Plots
using LaTeXStrings
using Measures # Required for setting margins

# 1. Definition of constants and points
const K = (3 - sqrt(5)) / 2
a_val = 1.0
b_val = 5.0
x_init = a_val + K * (b_val - a_val)

# Objective function (parabola)
f(t) = 0.4 * (t - 2.8)^2 + 1.5

# 2. Plot setup
p = plot(f, 0, 6,
    label = "",
    linewidth = 2,
    color = :royalblue,
    grid = true,
    gridalpha = 0.3,
    framestyle = :origin,
    size = (800, 500),
    xlims = (0, 6.5),
    ylims = (0, 6),
    bottom_margin = 15mm,
    top_margin = 10mm,
    left_margin = 5mm,
    right_margin = 5mm
)

# 3. Insert Equation as an annotation
equation = L"v = w = x = a + \left(\frac{3-\sqrt{5}}{2}\right)(b-a)"
annotate!(3.25, 5.8, text(equation, 16, :black, :center))


# 4. Helper function to draw stems (vertical lines with markers)
function draw_stem!(x, label_bottom)
    y = f(x)
    # Red vertical line
    plot!([x, x], [0, y], color=:red, linewidth=2, label="")
    # Red marker
    scatter!([x], [y], color=:red, markersize=7, markerstrokewidth=0, label="")
    
    # Bottom label (a, b)
    if !isempty(label_bottom)
        # Shift text slightly down (-0.5) to avoid overlap with the axis
        annotate!(x, -0.5, text(label_bottom, 16, :red, :center))
    end
end

# 5. Plot points a and b
draw_stem!(a_val, L"a")
draw_stem!(b_val, L"b")

# 6. Plot the initialization point (v, w, x)
# Draw the stem first
y_mid = f(x_init)
plot!([x_init, x_init], [0, y_mid], color=:red, linewidth=2, label="")
scatter!([x_init], [y_mid], color=:red, markersize=7, markerstrokewidth=0, label="")

# Manually stack letters above each other
annotate!(x_init, y_mid + 1.2, text(L"v", 16, :red, :center))
annotate!(x_init, y_mid + 0.8, text(L"w", 16, :red, :center))
annotate!(x_init, y_mid + 0.4, text(L"x", 16, :red, :center))

# Display the plot
display(p)

# Save figure (optional)
savefig("brent_initial_configuration.pdf")
```
