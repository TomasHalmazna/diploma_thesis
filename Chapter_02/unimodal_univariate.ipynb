Script that creates a graph of unimodal function of a single variable

```julia
using Plots

# Define a function that will be unimodal but with f(a)>f(x_0)<f(b) where x_0 is not minimum
f(x) = 0.5*(x + 2)^2 - 1

# Define the interval boundaries and the special point
a, b = -3, 1  # interval boundaries
x_0 = -1      # point where f(a)>f(x_0)<f(b) but not minimum

# Range for plotting
xs = range(-4, 2, length=400)

# Calculate function values at special points
f_a = f(a)
f_b = f(b)
f_x0 = f(x_0)

# Plot
plot(xs, f.(xs), linewidth=2, label="f(x)", color=:blue)

# Plot special points
scatter!([a], [f_a], color=:red, label="", markersize=8)
scatter!([b], [f_b], color=:red, label="", markersize=8)
scatter!([x_0], [f_x0], color=:black, label="\$x_0\$", markersize=8)

# Mark the interval [a,b]
y_interval = minimum(f.(xs)) - 0.5
plot!([a, b], [y_interval, y_interval], color=:black, lw=2, label="I = [a,b]")

# Draw vertical lines at both ends of the interval
plot!([a, a], [y_interval-0.2, y_interval+0.2], color=:black, lw=2, label="")
plot!([b, b], [y_interval-0.2, y_interval+0.2], color=:black, lw=2, label="")

# Add interval label
annotate!(((a+b)/2, y_interval - 0.4, text("I", 14, :center)))

# Add labels for a and b on x-axis
annotate!((a, y_interval - 1.2, text("a", 12, :center)))
annotate!((b, y_interval - 1.2, text("b", 12, :center)))

xlabel!("x")
ylabel!("f(x)")
```

```julia
# export the plot
png("unimodal_function_plot.png")
```
